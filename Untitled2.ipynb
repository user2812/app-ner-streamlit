```python
# 2. Chargement du corpus WikiANN (NER français)
dataset = load_dataset("wikiann", "fr", split="train[:1000]")

def reconstituer_phrase(ex):
    return " ".join(ex["tokens"])

phrases = [reconstituer_phrase(ex) for ex in dataset]
print(f"Nombre d'exemples chargés : {len(phrases)}")
print("\nExemple :\n", phrases[0])
```

```python
# 3. Chargement du modèle CamemBERT-ner (Hugging Face)
ner = pipeline("ner", model="Jean-Baptiste/camembert-ner", aggregation_strategy="simple")
```

```python
# 4. Extraction d'entités sur plusieurs textes
exemples = random.sample(phrases, 5)

for i, texte in enumerate(exemples, 1):
    print(f"\nTexte {i} :\n{texte}\n")
    entites = ner(texte)
    for ent in entites:
        print(f" - {ent['word']} ({ent['entity_group']}) - score: {ent['score']:.2f}")
```

```python
# 5. Statistiques sur les types d'entités
sample_texts = phrases[:100]
all_types = []

for txt in sample_texts:
    for ent in ner(txt):
        all_types.append(ent["entity_group"])

freq = Counter(all_types)
print("\nDistribution des types d'entités :", freq)

plt.figure(figsize=(6, 4))
plt.bar(freq.keys(), freq.values(), color='skyblue')
plt.title("Distribution des types d'entités")
plt.xlabel("Type")
plt.ylabel("Occurrences")
plt.grid(True)
plt.show()
```

```python
# 6. Évaluation quantitative NER avec alignement BIO
from sklearn.metrics import classification_report

tokenizer = AutoTokenizer.from_pretrained("Jean-Baptiste/camembert-ner")
label_list = dataset.features["ner_tags"].feature.names

def align_labels_with_tokens(tokens, labels, pred_entities):
    """
    Aligne les labels de référence et les labels prédits (BIO) pour chaque token.
    """
    pred_labels = ["O"] * len(tokens)
    for ent in pred_entities:
        ent_text = ent["word"]
        ent_type = ent["entity_group"]
        ent_tokens = ent_text.split()
        # Chercher la séquence d'ent_tokens dans tokens
        for i in range(len(tokens)):
            if tokens[i:i+len(ent_tokens)] == ent_tokens:
                pred_labels[i] = "B-" + ent_type
                for j in range(1, len(ent_tokens)):
                    pred_labels[i+j] = "I-" + ent_type
                break
    return pred_labels
y_true = []
y_pred = []

for ex in dataset.select(range(100)):
    tokens = ex["tokens"]
    true_labels = [label_list[idx] for idx in ex["ner_tags"]]
    text = " ".join(tokens)
    pred_entities = ner(text)
    pred_labels = align_labels_with_tokens(tokens, true_labels, pred_entities)
    y_true.extend(true_labels)
    y_pred.extend(pred_labels)

print("\nÉvaluation quantitative (BIO) sur 100 exemples :")
print(classification_report(y_true, y_pred, zero_division=0))
```

```python
# 📦 Imports
from transformers import pipeline, AutoTokenizer
from datasets import load_dataset
from sklearn.metrics import classification_report

# 📥 Chargement du dataset WikiANN (français)
dataset = load_dataset("wikiann", "fr", split="train[:1000]")
label_list = dataset.features["ner_tags"].feature.names

# 🤖 Chargement des deux modèles NER
ner_fr = pipeline("ner", model="Jean-Baptiste/camembert-ner", aggregation_strategy="simple")
ner_mbert = pipeline("ner", model="Davlan/bert-base-multilingual-cased-ner-hrl", aggregation_strategy="simple")

# 📌 Fonction d’alignement des prédictions avec les tokens
def align_labels_with_tokens(tokens, labels, pred_entities):
    pred_labels = ["O"] * len(tokens)
    for ent in pred_entities:
        ent_text = ent["word"]
        ent_type = ent["entity_group"]
        ent_tokens = ent_text.split()
        for i in range(len(tokens)):
            if tokens[i:i+len(ent_tokens)] == ent_tokens:
                pred_labels[i] = "B-" + ent_type
                for j in range(1, len(ent_tokens)):
                    pred_labels[i+j] = "I-" + ent_type
                break
    return pred_labels

# 🧪 Évaluation quantitative sur les 100 premiers exemples
y_true_fr, y_pred_fr = [], []
y_true_mbert, y_pred_mbert = [], []

for ex in dataset.select(range(100)):
    tokens = ex["tokens"]
    true_labels = [label_list[idx] for idx in ex["ner_tags"]]
    text = " ".join(tokens)

    # Modèle CamemBERT
    pred_fr = ner_fr(text)
    pred_labels_fr = align_labels_with_tokens(tokens, true_labels, pred_fr)
    y_true_fr.extend(true_labels)
    y_pred_fr.extend(pred_labels_fr)

    # Modèle BERT Multilingue
    pred_mbert = ner_mbert(text)
    pred_labels_mbert = align_labels_with_tokens(tokens, true_labels, pred_mbert)
    y_true_mbert.extend(true_labels)
    y_pred_mbert.extend(pred_labels_mbert)

# 📊 Affichage des résultats
print("📈 Évaluation Jean-Baptiste/camembert-ner :")
print(classification_report(y_true_fr, y_pred_fr, zero_division=0))

print("\n📈 Évaluation BERT Multilingue (Davlan/mbert) :")
print(classification_report(y_true_mbert, y_pred_mbert, zero_division=0))
```
